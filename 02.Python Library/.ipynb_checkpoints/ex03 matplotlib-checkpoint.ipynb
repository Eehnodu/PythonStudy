```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#    
x = np.arange(1,6)
np.random.seed(3)
y = np.random.randint(1,10,size=5)
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(y)
plt.show()
```

```python
plt.plot(x,y, marker = 'o', ms = 3, c = 'purple', mec='k', mew=7, mfc = 'w')
plt.show()
```

```python
plt.plot(x,y, ls = '-', color='b' ,marker='D', ms=10, mec='c',mew=4, mfc='y')
plt.plot(x,y, ls = '--', color='m', lw=5)
plt.plot(x,y, ls = '-.')
plt.plot(x,y, ls = ':')
plt.show()
```

```python
plt.subplot(3,2,1)
plt.plot(x,y, ls = '-', color='b' ,marker='D', ms=10, mec='c',mew=4, mfc='y')
plt.subplot(3,2,2)
plt.plot(x,y, ls = '--', color='m', lw=5)
plt.subplot(3,2,3)
plt.plot(x,y, ls = '-.')
plt.subplot(3,2,4)
plt.plot(x,y, ls = ':')
plt.subplot(3,2,5)
plt.plot(x,y, ls = ':')
plt.subplot(3,2,6)
plt.plot(x,y, ls = ':')
plt.show()
```

```python
#   
plt.rcParams['font.family'] = 'Malgun Gothic'

#    
test_x = np.arange(1,6)
np.random.seed(4)
test_y = np.random.randint(1,10,size=5)
plt.plot(test_x, test_y, label = 'test')
test_x1 = np.arange(1,6)
np.random.seed(10)
test_y1 = np.random.randint(1,10,size=5)
plt.plot(test_x1, test_y1, label = 'test1')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title("graph")
plt.grid()
plt.show()
```
